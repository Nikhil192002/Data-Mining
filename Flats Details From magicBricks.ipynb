```python
import requests
import pandas as pd
from bs4 import BeautifulSoup



url="https://www.magicbricks.com/property-for-rent/residential-real-estate?bedroom=2,3&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Service-Apartment,Residential-House,Villa&cityName=Bangalore"
r=requests.get(url)
print(r)

soup=BeautifulSoup(r.content,"html.parser")


box=soup.find_all("h2",class_="mb-srp__card--title")
flat=[]
for i in box:
    name=i.text.strip()
    flat.append(name)
print(len(flat))


box1=soup.find_all("div",class_="mb-srp__card__society")
socity=[]
for i in box1:
    s=i.text.strip()
    socity.append(s)
print(len(socity))
```

```python
box3=soup.find_all("div",class_="mb-srp__card__price--amount")
price=[]
for i in box3:
    p=i.text.strip()
    price.append(p)
print(len(price))
```

```python
box4=soup.find_all("div",class_="mb-srp__card__summary__list")
description=[]
for i in box4:
    d=i.text.strip()
    description.append(d)
print(len(description))
```

```python
box4=soup.find_all("ul",class_="mb-srp__card__kilometre__list")
near_By_distance=[]
for i in box4:
    n=i.text.strip()
    near_By_distance.append(n)
print(len(near_By_distance))
```

```python
socity_details=pd.DataFrame({"socity":socity,"Landmark":near_By_distance})
socity_details.head()
```

```python
flat_details=pd.DataFrame({"Flat":flat,"description":description,"Price":price})
flat_details
```
